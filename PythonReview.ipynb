```python
#If else 
tempreture = 50
if tempreture >= 70: #conditon:
    print("I do not need a jacket")
elif tempreture < 70 and tempreture >= 40: #other condition
    print("I need a jacket")
elif tempreture < 40:
    print("I need my winter coat")
else: #final condition if all others are false 
    print("I'm not leaving my house today anyway")
```

```python
#If-Else and For-Loops
grades = [85,65,99,50] #a list 
for grade in grades: #the for loops starts iterating through the list 
    if grade <= 100 and grade > 89: #conditions
        print( "A") #execution 
    elif grade <= 89 and grade >79:
        print ("B")
    elif grade <= 79 and grade > 69:
        print("C")
    elif grade <= 69 and grade > 59:
        print( "D")
    else:
        print( "F")
```

```python
#While Loops
x = 3
while x < 10: #The while runs until this condtion is false. Then it stops
    print(x)
    x = x+1
    
    
```

```python
#Data Cleaning
#Getting rid of spaces in column names 
#df = df.rename(columns = {"Old Name" : "New_Name"})
SSdf = SSdf.rename(columns = {"Product Name" : "Product_Name", "Product ID" : "Product_ID", "Postal Code" : "Postal_Code"})
```

```python
SSdf.head()
```

```python
#df.drop(columns = ["Column 1", "Column 2"])
SSdf2 = SSdf.drop(columns = "Ship Date")
SSdf2.head()
```

```python
#Changing data types 
#Changing dates to datetime format. This allows you to group numeric values by year and month. 

# df["Date Column"] = pd.to_datetime(pd.Series(df["Date Column]))

SSdf["Order Date"] = pd.to_datetime(pd.Series(SSdf["Order Date"]))
SSdf["Ship Date"]= pd.to_datetime(pd.Series(SSdf["Ship Date"]))
SSdf.head()
```

```python
#Analysis 
#Group By
#df.groupby("GroupCol")["AggCol"].agg()
#Show total sales by Region 
TotalSalesRegion = SSdf.groupby("Region")["Sales"].sum().round(2)
print(TotalSalesRegion)
```

```python
#Show total sales by year
TotalSalesYear = SSdf.groupby([SSdf["Order Date"].dt.year])["Sales"].sum().round(2)
#This takes your new date column and breaks into date parts
#df.groupby([df["Date Col"].dt.year])["Agg Col"].agg()
#You can also use month 
print(TotalSalesYear)
```

```python
#Pivot Tables 
#pd.pivot_table(data = the dataframe/table you're using, values = the values your going to aggregate 
#index = what you want on the rows side, 
#columns = what you want on the columns side
#aggfunc = the aggregrate function you're going to use

#Show total sales by region over the course of each year

TotalSalesPivot = pd.pivot_table(data = SSdf, values = "Sales", index = "Region", 
                                 columns = SSdf["Order Date"].dt.year, aggfunc =  'sum')
print(TotalSalesPivot) 
```

```python
TotalSalesPivot = TotalSalesPivot[[2011,2012,2013,2014]].round(2) #Round the values in your pivot table
print(TotalSalesPivot)
```

```python
#Subsetting Data
#newDf = df[Condtional statement], think of this like a Where Clause in SQL
SSSouthRegion = SSdf[SSdf["Region"] == "South"]
SSSouthRegion.head()


```
