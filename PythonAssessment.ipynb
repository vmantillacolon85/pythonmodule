10. Fasting blood sugar (fbs) is a blood glucose measurement, and a value greater than 120 means the patient is diabetic.  Create a pivot table with 'decade' as the index,  'fbs>1' as the values, 'sex' in columns, and mean as the aggregation function.  (This computes the likelihood of developing diabetes based on age and sex).  Then create a seaborn heat map from your pivot table, using any color palette you choose.  

```python
df.columns
```

```python
possible_diabetes = df.pivot_table(index = 'decade', columns = 'sex', values = 'fbs > 120', aggfunc = 'mean')
possible_diabetes
```

```python
sns.heatmap(possible_diabetes, cmap='mako')
```
